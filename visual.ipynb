```python
import h5py
import numpy as np
import gzip
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.colors import colorConverter
import nibabel as nib

case = '0008'
imslice = 105

# filename = "../data/Synapse/test_vol_h5/case0036.npy.h5"
# trainfile = "../data/Synapse/train_npz/case0026_slice100.npz"
gt = "../predictions/TU_Synapse224/TU_pretrain_R50-ViT-B_16_skip3_epo150_bs24_224/case" + case + "_gt.nii.gz"
img = "../predictions/TU_Synapse224/TU_pretrain_R50-ViT-B_16_skip3_epo150_bs24_224/case" + case + "_img.nii.gz"
pred = "../predictions/TU_Synapse224/TU_pretrain_R50-ViT-B_16_skip3_epo150_bs24_224/case" + case + "_pred.nii.gz"


image = nib.load(img)
im = np.array(image.dataobj)[:,:,imslice]
plt.imshow(im, interpolation='nearest')
plt.show()

ground = nib.load(gt)
gro = np.array(ground.dataobj)[:,:,imslice]
plt.imshow(gro, interpolation='nearest')
plt.show()

predict = nib.load(pred)
pre = np.array(predict.dataobj)[:,:,imslice]
plt.imshow(pre, interpolation='nearest')
plt.show()


```
